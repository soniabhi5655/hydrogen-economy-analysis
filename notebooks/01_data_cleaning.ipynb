```python
df = pd.read_excel('E:/Hydrogen-analysis/data/raw/IEA_Hydrogen_Projects.xlsx', sheet_name='Projects')
print(df.shape)
df.head()
```

```python
# See all column names
print(df.columns.tolist())
```

```python
# See data types and non-null counts
df.info()
```

```python
# See how many nulls in each column
df.isnull().sum()
```

```python
print(f"Total projects: {df.shape[0]}")
print(f"Total columns: {df.shape[1]}")
print(f"\nColumn names:\n{df.columns.tolist()}")
```

```python
# Keep only named columns
df = df.loc[:, ~df.columns.str.startswith('Unnamed')]

# Rename the messy column names (remove newlines and special characters)
df.columns = df.columns.str.replace('\n', ' ', regex=False)
df.columns = df.columns.str.strip()

print(df.columns.tolist())
print(df.shape)
```

```python
df = df.rename(columns={
    'Country (ISO-3)' : 'country',  
    'Project name'     : 'project_name',
    'Reference'        : 'reference',
    'Date online'      : 'date_online',
    'Decomission date' : 'decommission_date',
    'Status'           : 'status',
    'Technology'       : 'technology',
    'Product'          : 'product',
    'End use'          : 'end_use',
    'Announced Size'   : 'announced_size',
    'Normalised capacity' : 'capacity_normalized',
    'IEA zero-carbon estimated normalised capacity [Nm³ H₂/hour]' : 'capacity_zerocarbon',
    'Refs'             : 'refs',
    'Geographical details Decimal Degrees (Lat. N is positive, Long. E is positive)' : 'latitude'
})

print(df.columns.tolist())
```

```python
df.isnull().sum()
```

```python
print("Status values:\n", df['status'].value_counts())
print("\nTechnology values:\n", df['technology'].value_counts())
print("\nProduct values:\n", df['product'].value_counts())
```

```python
# Keep rows that at least have a country and status
df_clean = df.dropna(subset=['country', 'status'])

print(f"Rows before: {len(df)}")
print(f"Rows after:  {len(df_clean)}")
```

```python
df_clean['date_online'] = pd.to_numeric(df_clean['date_online'], errors='coerce')

# Check the year range we have
print(df_clean['date_online'].describe())
```

```python
df_clean.to_csv('E:/Hydrogen-analysis/data/processed/hydrogen_clean.csv', index=False)
print("Saved successfully!")
print(f"Final shape: {df_clean.shape}")
```
